```python
qc = QuantumCircuit(2)
initial_state = [1/np.sqrt(2), 1/np.sqrt(2)]
qc.initialize(initial_state, 0)
qc.cu1(np.pi/4, 0, 1)
qc.draw()
```

```python
statevector = execute(qc, Aer.get_backend('statevector_simulator')).result().get_statevector()
plot_bloch_multivector(statevector)
```
